```python
# Import all necessary libraries and root drive
import os
os.chdir("/content/drive/MyDrive/Pengolahan Citra")
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage

from skimage.io import imread, imshow
```

# Filtering

```python
img_path = './taylor.jpg'

# convert image to grayscale
taylor = imread(img_path, as_gray=True)

# Plot the image with histogram
plt.imshow(np.array(taylor))
plt.colorbar()
```

## MEAN Filter

```python
# initialize the filter of size 5 by 5
# the filter is divided by 25 for normalization
k = np.ones((5,5))/25
print(k)
```

```python
np.sum(k)
```

```python
# Performing convolution
b = scipy.ndimage.filters.convolve(taylor, k)

plt.imshow(b)
plt.show()

## Mean Filter remove noise but blurred the image during the process
```

## MEDIAN Filter

```python
# Median filter
med = scipy.ndimage.filters.median_filter(taylor, size=5, footprint=None, output=None, mode='reflect', cval=0.0, origin=0)

plt.imshow(med)
plt.show()

## Median filter is commonly used to remove noise and impulse or black and white spot noise
```

## MAX Filter

```python
# Max Filter
max = scipy.ndimage.filters.maximum_filter(taylor, size=5, footprint=None, output=None, mode='reflect', cval=0.0, origin=0)

plt.imshow(max)
plt.show()

## Max filter enhances bright points, suitable for BW image
## the white pixels grew after applying the filter
```

## MIN Filter

```python
# Min Filter
min = scipy.ndimage.filters.minimum_filter(taylor, size=5, footprint=None, output=None, mode='reflect', cval=0.0, origin=0)

plt.imshow(min)
plt.show()

## This filter is used to enhance the darkest points in the image
```

# Edge Detection

```python
from skimage import filters, feature

# Sobel Filter
sob = filters.sobel(taylor)

plt.imshow(sob)
plt.show()
```

```python
# Prewitt Filter
prew = filters.prewitt(taylor)

plt.imshow(prew)
plt.show()
```

```python
# Canny Filter
can = np.asarray(taylor).astype(float)

np.min(can), np.max(can)

can = can/float(np.max(can))
can = feature.canny(can, sigma=1, low_threshold=0.05, high_threshold=0.24)

plt.imshow(can)
plt.show()
```

```python
# Laplace Filter
lap = scipy.ndimage.filters.laplace(taylor)

plt.imshow(lap)
plt.show()
```

```python
# Plot all the figures

fig = plt.figure(constrained_layout=True)

ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
ax3 = fig.add_subplot(233)
ax4 = fig.add_subplot(234)
ax5 = fig.add_subplot(236)

ax1.imshow(taylor)
ax1.set_title('Original Image')
ax2.imshow(sob)
ax2.set_title('Sobel Filter')
ax3.imshow(prew)
ax3.set_title('Prewitt Filter')
ax4.imshow(can)
ax4.set_title('Canny Filter')
ax5.imshow(lap)
ax5.set_title('Laplace Filter')

for ax in [ax1, ax2, ax3, ax4, ax5]:
    ax.set_aspect('equal')

plt.show()
```
